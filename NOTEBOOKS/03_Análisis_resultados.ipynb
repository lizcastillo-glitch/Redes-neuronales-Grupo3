```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from io import StringIO

sns.set(style="whitegrid")
```

```python
# URL RAW del CSV en GitHub (YA LA TIENES)
GITHUB_RAW_URL = "https://raw.githubusercontent.com/lizcastillo-glitch/Redes-neuronales-Grupo3/main/data/synthetic_data.csv"

response = requests.get(GITHUB_RAW_URL)

if response.status_code != 200:
    raise FileNotFoundError("❌ No se pudo descargar synthetic_data.csv desde GitHub")

df = pd.read_csv(StringIO(response.text))

print("✅ Datos cargados correctamente desde GitHub")
df.head()
```

```python
print("Dimensiones del dataset:", df.shape)
print("\nDistribución de clases:")
df["label"].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="label", data=df)
plt.title("Distribución de clases de sentimiento")
plt.xlabel("Clase")
plt.ylabel("Frecuencia")
plt.show()
```

### Análisis de Resultados

La distribución de clases muestra un conjunto de datos pequeño y balanceado,
diseñado con fines académicos para validar el correcto funcionamiento de la red neuronal.

Dado el tamaño reducido del dataset, los resultados obtenidos en los experimentos
deben interpretarse como una prueba de concepto y no como un modelo listo para producción.

Este enfoque permite evaluar la correcta implementación del forward y backward propagation,
así como el impacto de distintas funciones de activación y arquitecturas.


### Conexión con el Proyecto Final

La implementación manual de la red neuronal permite comprender en profundidad
las limitaciones computacionales y de escalabilidad del enfoque desde cero.
